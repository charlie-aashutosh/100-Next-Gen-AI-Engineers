```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/fraud_transactions_synthetic.csv")
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
#the whole data gets divided into Train Set [Make the Model Learn] and Test Set [Validate the performance]
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
import tensorflow as tf
from tensorflow import keras #keras is a module within Tensorflow Library
from tensorflow.keras import layers #The NN will be build as a Sequential Layer of Input, Hidden and Output Layers
```

From the given dataset, is_fraud is the target variable or we can say the output dependent variable and all the other columns are input independent variables

```python
x=data.drop(['is_fraud'],axis=1) #'x' holds all the input independent columns
y=data['is_fraud'] #'y' holds only the output dependent column
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.85)
```

- Whenever we are training the model it is important that we provide details about the input (x_train) as well as identify what the output looks like (y_train) because at this stage, the model is learning under supervision (Supervised ML) - that's why it is important that the model has access to both input as well as output parameters

- Whenever we are testing the performance of the model, it is important that we test the predicted output (y_pred) from input (x_test) against the actual values (y_test)


```python
model=keras.Sequential(
    [
        layers.Input(shape=(x_train.shape[1],),name='input'),
        layers.Dense(32,activation='relu',name='hidden_layer_1'),
        layers.Dense(16,activation='relu',name='hidden_layer_2'),
        layers.Dense(1,activation='sigmoid',name='output_layer')
    ]
)
```

```python
model.compile(
    optimizer='adam',
    loss=keras.losses.BinaryCrossentropy(name='binary_cross_entopy'),
    metrics=[
        keras.metrics.BinaryAccuracy(name='Accuracy'),
        keras.metrics.MeanSquaredError(name='MSE')
    ]
)
```

```python
model.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test))
```
