```python
#Visualizing the distribution of Monthly Data Usage
plt.figure(figsize=(6,4))
sns.histplot(data['monthly_data_usage_gb'],bins=30) #bins we manage the thickness of the Histogram
plt.title("Distribution of Monthly Data Usage")
plt.xlabel("Data Usage (GB)") #on the x-axis
plt.ylabel("Number of Customers") #on the y-axis
plt.show()
```

#Finance Domain

```python
data=pd.read_csv("/content/finance_dataset.csv")
```

```python
data.head()
```

- Here Credit Score is the Risk Indicator and high Credit Score means lower chances of Defaulting
- Higher income gives access to larger loan amounts
- Existing Loan is the debt burden on the Customer and if Existing Load is higher then chances of Defaulting also increases to a higher %

- Q. What is the maximum amount of Loan issued ?
- Q. Find the median income of the Customers
- Q. How many Customers have defaulted so far ?


```python
#Income distribution
plt.figure(figsize=(6,4))
sns.histplot(data['annual_income'],bins=30) #bins we manage the thickness of the Histogram
plt.title("Distribution of Income")
plt.xlabel("Annual Income") #on the x-axis
plt.ylabel("Number of Customers") #on the y-axis
plt.show()
```

```python
#Credit Score distribution
plt.figure(figsize=(6,4))
sns.histplot(data['credit_score'],bins=30) #bins we manage the thickness of the Histogram
plt.title("Distribution of Credit Score")
plt.xlabel("Credit Score") #on the x-axis
plt.ylabel("Number of Customers") #on the y-axis
plt.show()
```

```python
#Relationship between Credit Score and Loan Amount
plt.figure(figsize=(6,4))
sns.scatterplot(
    x=data['credit_score'],
    y=data['loan_amount']
)
plt.title("Credit Score vs Loan Amount")
plt.xlabel("Credit Score")
plt.ylabel("Loan Amount")
plt.show()
```
